```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
ds_o = xr.open_dataset("19000101.ocean_daily.nc", decode_times=False)

ds = xr.open_dataset("../OM4_05/19000101.ocean_daily.nc", decode_times=False)
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(facecolor=(0.8, 0.8, 0.8))
pc = (ds_o['tos'] - ds['tos']).plot(vmin=-0.03, vmax=0.03, cmap="RdBu_r")
pc.colorbar.set_label("Surface cooling [C] due to one day of Freddie's idealized OTEC")
plt.ylabel("latitude [degrees]")
plt.xlabel("longitude [degrees]")
plt.savefig("figures/global_otec_test.png", bbox_inches="tight", dpi=150)
```
